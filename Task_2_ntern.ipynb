```python
# Importing libraries
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
# Step 1: Load dataset
# If you have the dataset locally, replace 'titanic.csv' with your file path
# Example dataset from ydata-profiling GitHub repo (Titanic dataset)
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

df = pd.read_csv(url)

```

```python
# Step 2: Generate the profiling report
profile = ProfileReport(df, title='Titanic Dataset EDA Report', explorative=True)
```

```python
# Step 3: Save the report as an HTML file
profile.to_file("titanic_eda_report.html")


```
